**Conversión de Datos**

Así como se sabe, los datos pueden venir en el formato incorrecto, y convertirlos correctamente es fundamental.
- La instrucción input() simpre convierte la entrada del usuario en una cadena de texto, sin importar el tipo de dato ingresado.
- Se pueden converitr datos de un tipo a otro para solucionar problemas de calidad de datos.
- La instrucción int() convierte cualquier tipo de valor en un número entero.
- La instrucción float() convierte valores en flotantes.
- Los valores se pueden converitr en cadenas de texto con la instrucción str().
- Las instrucciones int(), float(), y str() son ejemplos de conversión explícita, que son realizadas por instrucción de un programador.

```python
x = "55"
print(type(x))
y = int(x)
print(type(y))

altura = int(input())
print(altura)
print(type(altura))

a = 15
b = float(a)
print(b)
```

**Operadores de Comparación**

Las operaciones de comparación son clave para el desarrollo de programas.
- En la afirmación 5 < 9, ¿5 es menor a 9?
- Una operación de comparación solo tiene dos resultados: Sí o No-
- Por ejemplo, los circuitos electrónicos dentro de las computadoras utilizan millones de interruptores pequeños para almacenar valores lógicos True/False.
- TRUE --> 1; FLASE --> 0; Número representativos del código binario.
- El booleano es un tipo de dato que solo tiene dos posibles valores : Verdadero (TRUE) o Falso (FALSE)
- Las operaciones de comparación y los valores boolenaos permiten que las máquinas toman decisiones.

```python
print(30<25)
print(5<9)
print(50>100)


soil_moisture = 80
soil_moisture = 120
print(soil_moisture < 100)
```

**Operaciones Lógicas**

Los computadores modernos pueden realizar tareas complejas rápidamente porque estás se pueden descomponer en muchas operaciones más simples.
- Las operaciones lógicas son necesarias para que las máquinas evalúen escenarios complejos.
- La **operación AND** resulta en un valor verdadero solo cuando sus entradas son verdaderas al mismo tiempo.
- La **operación OR** resulta en un valor Verdadero si al menos una de sus entradas es verdadera.
- Las operaciones lógicas pueden insertarse en las instrucciones print().


```python
True and True
True and False
False and False

True or True
True or False
#False or False
```
